```python
from bing_image_downloader import downloader

classes = {
    "no_damage": "car images",
    "minor_damage": " minor car damage",
    "major_damage": " major car accident damage"
}

for folder, query in classes.items():
    downloader.download(
        query,
        limit=20,
        output_dir="car_damage",
        adult_filter_off=True,
        force_replace=False,
        timeout=60
    )
```
